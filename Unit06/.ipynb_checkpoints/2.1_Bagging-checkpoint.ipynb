```python
from sklearn.ensemble import BaggingClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=datasets.load_iris()
X=iris.data
y=iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
len(X_test)
```

```python
from sklearn import tree
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
clf=tree.DecisionTreeClassifier()
# clf=svm.LinearSVC()
# clf=GaussianNB()
```

```python
bagging=BaggingClassifier(estimator=clf,n_estimators=10,
                          bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
# estimator分類器 n_estimators生成分類器的數量  bootstrap=True 取後放回 false取厚不放回
```

```python
bagging.fit(X_train,y_train)
```

```python
bagging.predict(X_test)
```

```python
bagging.score(X_train,y_train)
```

```python
bagging.score(X_test,y_test)
```

```python
plt.scatter(X[:,2],X[:,3],c=y)
```

```python
plt.scatter(X[:,2],X[:,3],c=bagging.predict(X))
```

```python
clf=tree.DecisionTreeClassifier()
bagging=BaggingClassifier(estimator=clf,n_estimators=10,
                          bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
bagging.fit(X_train,y_train)
print("bagging.predict(X_test) = ", bagging.predict(X_test))
print("-"*100)
print("bagging.score(X_train,y_train) = ",bagging.score(X_train,y_train))
print("-"*100)
print("bagging.score(X_test,y_test) = ",bagging.score(X_test,y_test))
plt.scatter(X[:,2],X[:,3],c=y)
plt.scatter(X[:,2],X[:,3],c=bagging.predict(X))
```

```python
clf=svm.LinearSVC()
bagging=BaggingClassifier(estimator=clf,n_estimators=10,
                          bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
bagging.fit(X_train,y_train)
print("bagging.predict(X_test) = ", bagging.predict(X_test))
print("-"*100)
print("bagging.score(X_train,y_train) = ",bagging.score(X_train,y_train))
print("-"*100)
print("bagging.score(X_test,y_test) = ",bagging.score(X_test,y_test))
plt.scatter(X[:,2],X[:,3],c=y)
plt.scatter(X[:,2],X[:,3],c=bagging.predict(X))
```

```python
clf=GaussianNB()
bagging=BaggingClassifier(estimator=clf,n_estimators=10,
                          bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
bagging.fit(X_train,y_train)
print("bagging.predict(X_test) = ", bagging.predict(X_test))
print("-"*100)
print("bagging.score(X_train,y_train) = ",bagging.score(X_train,y_train))
print("-"*100)
print("bagging.score(X_test,y_test) = ",bagging.score(X_test,y_test))
plt.scatter(X[:,2],X[:,3],c=y)
plt.scatter(X[:,2],X[:,3],c=bagging.predict(X))
```
