```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [-1, -4.3, 15, 21, 31]
plt.plot(data)
plt.show()
```

```python
plt.plot(data, 'o--b')  # o是點點 --式虛線 b是藍色
plt.show()
```

```python
days = range(1, 7)
celsius_min = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
celsius_max = [28.6, 24.2, 22.2, 31.3, 27.5, 30.5]
plt.xlabel('days')
plt.ylabel('。c')
# 出現方框
plt.grid(True)
# 限制xy軸範圍
xmin, xmax, ymin, ymax = 0.5, 6.5, 16, 32.5
plt.axis([xmin, xmax, ymin, ymax])
plt.plot(days, celsius_min, 'g--s',
         days, celsius_max, 'b-o')
plt.show()
```

```python
# 繪製 sin cos圖
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x , sinus, x, cosinus)
plt.show()
```

```python
# 繪製sin()分布圖
x = np.linspace(0, 2*np.pi, 50)
sinus = np.sin(x)
plt.scatter(x, sinus)
plt.title("sin")
plt.show()
```

```python
labels = ["Python","C++","Java","JS","C","C#"]
index = np.arange(len(labels))
ratings = [5.16, 5.73, 14.99, 3.17, 11.86, 4.45]
change = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]

plt.bar(index, ratings)
plt.xticks(index, labels)
plt.ylabel("usage")
plt.title("Programming language usage")
plt.show()
```

```python
x= [21,42,23,4,5,26,77,88,9,10,31,32,33,
34,35,36,37,18,49,50,100]
num_bins = 5
# n每個區間有幾筆資料  bins是每一條直線的位置
n, bins, patches = plt.hist(x, num_bins) #num_bins 為要分割的區間數量
print(n)
print(bins)
print(patches)
plt.show()
```

```python
# 圓餅圖
labels = ["Python","C++","Java","JS","C","C#"]
ratings = [5, 6, 15, 3, 12, 4]
plt.pie(ratings, labels=labels)
plt.title("Programming language usage")
plt.axis("equal")
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(4,2,1)
plt.plot(x, sinus, "r-o")
plt.subplot(4,2,2)
plt.plot(x, cosinus, "g--")
plt.subplot(4,2,3)
plt.plot(x, sinus, "b-o")
plt.subplot(4,2,4)
plt.plot(x, cosinus, "y--")
plt.show()
```

```python
# 多軸圖表
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
sinhs = np.sinh(x)
fig, ax = plt.subplots()
ax.plot(x, sinus, "r-o")
ax2 = ax.twinx()
ax2.plot(x, sinhs, "g--")
plt.show()
```
