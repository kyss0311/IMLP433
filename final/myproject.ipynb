### 皮爾遜相關係數(coefficient)

```python
def coefficient(x, y):
    
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    # print(x_mean)
    # print(y_mean)
    
    xi_x = [v - x_mean for v in x]
    yi_y = [v - y_mean for v in y]
    
    dataxy =[0]*len(x)
    datax =[0]*len(x)
    datay =[0]*len(x)
    
    for i in range(len(x)):
        dataxy[i] = xi_x[i]*yi_y[i]
        datax[i] = xi_x[i]**2
        datay[i] = yi_y[i]**2
    
    txy = np.sum(dataxy)
    tx = np.sum(datax)
    ty = np.sum(datay)
      
    r = txy/((tx**0.5)*(ty**0.5))
    print(f"皮爾遜相關係數(coefficient) = {r}")
```

```python
for i in df2.columns[1:]:
    if i != "Calories":
        print(f"{i}跟Calories")
        coefficient(df2[i],df2.Calories)
```

這裡經過計算皮爾遜相關係數可以得知上面的推測，身高體重跟卡路里有相關是錯誤的

## 評估模型
MAE（Mean Absolute Error）抵銷正負誤差的方式，除了平方之外，還有取絕對值，MAE就是取絕對值來計算平均誤差，相對之下對極值比較不敏感 <br>
MSE（Mean Square Error）MSE主要以平方來避免誤差正負的互相抵銷，但也因為平方的特性，所以當單一bias大的時候會有懲罰作用，也就是說MSE對於極值會相對敏感<br>
RMSE（Root Mean Square Error）RMSE主要就是MSE拿去取根號，取根號的目的是讓他與y的單位變得一致，所以解釋起來會比較直觀

```python
#載入迴歸常見的評估指標套件
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_train, Y_train)
MSE = metrics.mean_squared_error(y_train, Y_train)
RMSE = metrics.root_mean_squared_error(y_train, Y_train)

print("MAE =  ", MAE)
print("MSE = ", MSE)
print("RMSE = ", RMSE)
```

### 比較Ridge Lasso
兩者皆會將具有相關性的變數推向彼此並避免使的模型參數中出現一個有極大正係數與另一個極大負係數的一起出現的情況 <br>
Ridge 會將不相干變數的迴歸係數會被逼近為0(不會剛好等於0)，我們可以藉此降低資料集中的雜訊，但是最後的模型可能還是會存在一些不重要的參數，多多少少還是會影響模型最後的正確性<br>
Lasso 會將不具有影響力的迴歸係數變成0，等於可以自動進行Feature selection，但會犧牲掉模型的正確性

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
def ridge(n):
    rr = Ridge(alpha=n)
    rr.fit(X_train, y_train)
    
    Ridge_train_score = rr.score(X_train,y_train)
    Ridge_test_score = rr.score(X_test, y_test)
    
    print(f"ridge regression train score alpha = {n}:", Ridge_train_score)
    print(f"ridge regression test score alpha = {n}:", Ridge_test_score)
    print('-'*100)

    return np.ravel(rr.coef_)
    

r1 = ridge(1)
r001 = ridge(0.01)
r100 = ridge(100)

plt.figure(figsize=(6, 4))
plt.plot(r1,alpha=0.7,linestyle='none',marker='*',markersize=5,color='red',label=r'Ridge; $\alpha = 1$')
plt.plot(r001,alpha=0.7,linestyle='none',marker='d',markersize=5,color='blue',label=r'Ridge; $\alpha = 0.01$')
plt.plot(r100,alpha=0.7,linestyle='none',marker='o',markersize=5,color='green',label=r'Ridge; $\alpha = 100$')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient Magnitude')
plt.xticks(ticks=range(6),labels=['Age', 'Height', 'Weight', 'Duration','Heart_Rate','Body_Temp'])
plt.legend(fontsize=8,loc=1)
plt.show()
```

結論：Ridge 的 alpha 係數越地評分越高

```python
def lasso_(n):
    lasso = Lasso(alpha=n, max_iter=1000000)
    lasso.fit(X_train,y_train)
    train_score=lasso.score(X_train,y_train)
    test_score=lasso.score(X_test,y_test)
    coeff_used = np.sum(lasso.coef_!=0)
    print(f"training score for alpha= {n}:", train_score )
    print(f"test score for alpha = {n}: ", test_score)
    print(f"number of features used: for alpha = {n}:", coeff_used)
    print('-'*100)

    return lasso.coef_
    
la1 = lasso_(1)
la001 = lasso_(0.01)
la00001 = lasso_(0.0001)

plt.figure(figsize=(6, 3))
plt.xlabel('Coefficient Index',fontsize=10)
plt.ylabel('Coefficient Magnitude',fontsize=10)
plt.xticks(ticks=range(6),labels=['Age', 'Height', 'Weight', 'Duration','Heart_Rate','Body_Temp'])
plt.legend(fontsize=8,loc=3)

plt.plot(la1,alpha=0.7,linestyle='none',marker='*',markersize=6,color='red',label=r'Lasso; $\alpha = 1$',zorder=7) 
plt.plot(la001,alpha=0.5,linestyle='none',marker='d',markersize=6,color='blue',label=r'Lasso; $\alpha = 0.01$') 
plt.plot(la00001,alpha=0.8,linestyle='none',marker='v',markersize=6,color='green',label=r'Lasso; $\alpha = 0.0001$') 
plt.legend(fontsize=8,loc=3)
plt.show()
```

結論：Lasso 的 alpha 係數越地評分越高 但就沒有把較不相關的係數去掉

## 定義一個可以評分的函式

```python
def mypredict(clf,X_test, y_test):
    pred=clf.predict(X_test)
    error_index = []
    for i in range(len(pred)):
        if [pred[i]] != y_test.values[i]:
            error_index.append(i)
    
    print('accuracy:',1-len(error_index)/len(X_test))
    print('number of errors :%d/%d'%(len(error_index),len(X_test)))
    print('-'*100)
    print('error_index :',error_index)
    print('miss classification :',pred[error_index])
    data_index=X_test.iloc[error_index].index
    print('error data index:',data_index)
```

```python
#由於KNN分類法是藉由確認最近的觀察值來預測分類，因此變數的比例就會影響很大
#載入標準化比例尺(StandardScaler)套件，只將所有`特徵值`做標準化
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df2.drop(["Gender", "Calories", "Heavy_level", "Burn_level"], axis=1))
scaled_features = scaler.transform(df2.drop(["Gender", "Calories", "Heavy_level", "Burn_level"], axis=1))
df_feat = pd.DataFrame(scaled_features, columns=df2.columns[1:-3])
print(df_feat)
```

```python
X = df_feat
y = df2["Heavy_level"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

precision(可信度) = TP/(TP+FP) , recall(多少正確的被檢測出來)= TP/(TP+FN)  , f1-score(模型的指標) = 2*(precision*recall) / (precision+recall)

```python
#測試KNN演算法的好壞
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
print('-'*100)
print(classification_report(y_test,pred))
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(confusion_matrix(y_test,pred), annot=True, fmt='d', cmap='PuBu')  #  fmt='d'印出整書註解  cmap='PuBu'設定顏色
```

```python
error_rate = []
best_k =0
for i in range(1,50):
    knn_i = KNeighborsClassifier(n_neighbors=i)
    knn_i.fit(X_train,y_train)
    pred_i = knn_i.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    if np.mean(pred_i != y_test) == min(error_rate):
        best_k = i
```

```python
print("錯誤率最低的K = ", best_k, ",錯誤率 = ", min(error_rate))
```

```python
#將k=1~60的錯誤率製圖畫出。k=23之後，錯誤率就在5-6%之間震盪。
plt.figure(figsize=(8,5))
plt.plot(range(1,50),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

#### 視覺化分群的過程

```python
import scipy.cluster.hierarchy as sch
dis = sch.linkage(data,metric = 'euclidean',method='ward')
sch.dendrogram(dis)
plt.title('Hierarchical Clustering')
plt.show()
```

## 使用Scipy進行分群
metric: 距離的計算方式
method: 群與群之間的計算方式，"single", "complete", "average", "weighted", "centroid", "median", "ward"

#### 以距離分群 經過測試出來1000左右會分成三群 但不符合一開始用BMI分的結果

```python
max_dis = 1000
clusters = sch.fcluster(dis,max_dis,criterion='distance')
plt.scatter(df['Height'],df['Weight'],c=clusters)
```

#### 以群數分群

```python
K = 3
clusters = sch.fcluster(dis,K,criterion='maxclust')
plt.scatter(df['Height'],df['Weight'],c=clusters)
```
